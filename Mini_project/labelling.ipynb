```python
# Install dulu: pip install transformers torch scipy

from transformers import pipeline
import pandas as pd

# 1. Load Data
df = pd.read_csv('dataset_tugas_purbaya_vs_srimulyani_lengkap.csv')

# 2. Siapkan Model Sentimen Bahasa Indonesia
# Model ini cukup populer untuk teks Indonesia
pretrained_name = "w11wo/indonesian-roberta-base-sentiment-classifier" 
sentiment_pipeline = pipeline("sentiment-analysis", model=pretrained_name, tokenizer=pretrained_name)

# 3. Fungsi Labeling
def get_sentiment(text):
    try:
        # Truncate text jika terlalu panjang (>512 token)
        result = sentiment_pipeline(text[:512])[0]
        return result['label'], result['score']
    except Exception as e:
        return "ERROR", 0.0

# 4. Terapkan ke data (Mungkin lama jika data banyak)
# Coba dulu ke 100 data pertama
df_sample = df.head(100).copy()
df_sample[['predicted_label', 'confidence']] = df_sample['text'].apply(lambda x: pd.Series(get_sentiment(x)))

print(df_sample[['text', 'predicted_label']].head())
```
